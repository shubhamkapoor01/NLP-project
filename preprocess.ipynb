```python
from sklearn.datasets import fetch_20newsgroups
```

```python
fetched_data = fetch_20newsgroups()
```

```python
import numpy as np
```

```python
raw_text = fetched_data.data[:5]
raw_text
```

```python
cleaned_text = []

def toLowerCase(raw_text):
  for words in raw_text:
    cleaned_text.append(str.lower(words))

toLowerCase(raw_text)
```

```python
cleaned_text[:5]
```

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
```

```python
# WORD TOKENIZE

cleaned_text_2 = [word_tokenize(i) for i in cleaned_text]
cleaned_text_2[:5]
```

```python
# REMOVE SPECIAL CHARACTER

import re

cleaned_text_3 = []

for words in cleaned_text_2:
  clean = []
  for w in words:
    res = re.sub(r'[^\w\s]', "", w)
    if res != "":
      clean.append(w)
    cleaned_text_3.append(clean)

cleaned_text_3[:5]
```

```python
# STOPWORD REMOVAL

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

cleaned_text_4 = []

for words in cleaned_text_3:
  w = []
  for word in words:
    if not word in stopwords.words('english'):
      w.append(word)
    cleaned_text_4.append(w)

cleaned_text_4[:5]
```

```python
# STEMMING

from nltk.stem.porter import PorterStemmer
port = PorterStemmer()

cleaned_text_5 = []

for words in cleaned_text_4:
  w = []
  for word in words:
    w.append(word)
  cleaned_text_5.append(w)

cleaned_text_5[:5]
```

```python
# LEMITIZATION

from nltk.stem.wordnet import WordNetLemmatizer
wnet = WordNetLemmatizer()
import nltk
nltk.download('wordnet')

lem = []

for words in cleaned_text_4:
  w = []
  for word in word:
    w.append(wnet.lemmatize(word))
  lem.append(w)

lem[:5]
```
